### B. Data Visualization (matplotlib)

```python
df.loc[(df.Continent == 'Asia') & (df.Region == 'Southern Asia'),years].head().plot(kind='line')
```

```python
haiti = df.loc['Haiti',years]
haiti
```

```python
# revamp pake matplotlib
haiti.index = haiti.index.map(int)
haiti.plot()

plt.title('Immigration from Haiti')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
```

```python
# berikan anotasi/catatan di lonjakan tahun 2010 ketika Canda meningkatkan uopaya
haiti.index = haiti.index.map(int)
haiti.plot()

plt.title('Immigration from Haiti')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
plt.text(2000,6000,'2010 Earthquake')
```

Contoh soal 2:  
Compare the trend of top 5 countries that contributed the most to immigration Canada

```python
df.columns
```

```python
df.sort_values('Total', ascending=False).head()[years].T.plot(figsize=(14.8)) #ascending kalo true dari yg plt kecil, kalo false dr yg gede
plt.title('Immigration Trend of the Top 5 Countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')

plt.show()
```

Contoh soal no 3: 

Compare the number of Iceland immigrants (country = 'Iceland') to Canada from 1980 to 2013

```python
df.loc['Iceland',years].plot(kind='barh', figsize=(10,6))

plt.title('Iceland Immigrants to Canada from 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
```
