### Histogram

Histograms --> representing frequency distribution out of a numerical dataset  

How it works in py -->   

1. partitioning the x axis into several bins, menetapkan setiap data point dalam kumpulan data kt ke sebuah bin, dan kemudian menghitung jumlah data point yg telah ditetapkan ke setiap bin  

2. y axis = frequency (banyaknya data point in every bins)  
(kita dpt mengubah ukuran bin & biasanya kt perlu menyesyaikan agar dist. ditampilkan dgn baik)

Contoh soal lg:  

What is the frequency distribution of the num (population) of new immigrants from the various countries to Canada in 2013?

```python
# pertama, dapatkan range bin & jumlah frekuensi
df['2013'].head()
```

```python
count, bin_edges = np.histogram(df['2013'])

print(count) #to find out frequency count
print(bin_edges) #to find out bin ranges, default = 10 bins
```

```python
df['2013'].plot(kind='hist', figsize =(8,5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
```

label sumbu x tidak cocok dgn ukuran bin --> diperbaiki dgn xticks

```python
df['2013'].plot(kind='hist', figsize =(8,5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
```

Contoh soal lg:  

What is the immigration dist. for Denmark, Norway & Sweden for years 1980-2013?

```python
df.loc[['Denmark', 'Norway', 'Sweden'], years].T.plot(kind='hist', figsize=(10,6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

to improve impact & esthetics:  
* increase the bin size to 15 by passing in bins parameter  
* set transparency to 60% by passing in alpha parameter
* label the x-axis by passing in x-label parameter
* change the colors of the plots by passing in color parameter


```python
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].T
# get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

#unstacked histo
df_t.plot(kind='hist',
           figsize=(10,6),
           bins=15,
           alpha=0.6,
           xticks=bin_edges,
           color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

jika tidak ingin plot saling tumpang tindih, dapat ditumpuk dgn parameter stacked.  

mari juga menyesuaikan label x min & maks utk menghilangkan celah extra di tepi plot.  

kita bs pass tuple (min,max) dgn parameter xlim:

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges [0] - 10 # 1st bin value is 31.0, adding buffer of 10 for aesthetic purposes
xmax = bin_edges [-1] + 10 # last bin value is 308.0, adding buffer of 10 for aest purposes

# stacked Histo
df_t.plot(kind='hist',
           figsize=(10,6),
           bins=15,
           xticks=bin_edges,
           color=['coral', 'darkslateblue', 'mediumseagreen'],
           stacked=True,
           xlim=(xmin, xmax))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

Plot di atas menunjukkan jumlah total imigran yg dikelompokkan setiap tahun. Kita bisa melihat dgn jelas dampak dari krisis keuangan; jumlah imigran ke Kanada mulai meingkat pesat setelah 2008

memberi anotasi pada plot dgn method annotate pada scripting layer atau antarmuka pyplot. Parameter:

* s: str, the text of annotation
* xy: tuple specifying the (x,y) point to annotate (in this case, end point of arrow)
* xytext: tuple specifying the (x,y) point to place the text (start point of arrow)
* xycoords: the coordinate system that xy is given in -'data' uses the coord. system of the obj. being annotated(default)
* arrowprops: takes a dictionary of properties to draw the arrow

- arrowstyle: specifies the arrow style, '->' is the std arrow
- connectionstyle: specifies the connection type.arc3 is a straight line
- color: specifies arrow color
- lw: specifies the line width

also annotate pada teks untuk berada di atas panah, parameter additional:

* rotation: rotation angle of text in degress (counter clockwise)
* va: vertical alignment of text ['center' | 'top'| 'bottom' | 'baseline']
* ha: horziontal alignment of text ['center' | 'right'| 'left']

```python
df_ice = df.loc['Iceland',years]
```

```python
df_ice.plot(kind='bar', figsize=(10,6), rot=90)

plt.title('Iceland Immigrants to Canada from 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

# Annotate arrow
plt.annotate('',  # s: str, leave blank for no text
             xy =(32,70), # place head of arrow at point (year 2012, pop 70)
             xytext=(28,20), #place base of arrow at point (year 2008, pop 20) 
             xycoords='data',
             arrowprops=dict(arrowstyle='->',
             connectionstyle='arc3',
             color='blue',
             lw=2))

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', 
              xy=(28,30), # start text at pt (year 2008, pop 30)
              rotation=72.5, # trial & error
              va='bottom',
              ha='left' )
```

### Pie Charts


```python
df.drop(columns=['Region', 'DevName'], inplace=True)
```

```python
# step 1: gather & group data
df_cons = df.groupby('Continent', axis=0).sum() #hrs sll ad funct kek sum
df_cons.head()
```

```python
# step 2: plot, additional parameters:
```

* autopct = a str or func -> label the wedges w numeric value. label will be placed inside the wedge. if it's format string, label will be fmt%pct
* startangle - rotates the start of the pie chart by angle degrees counterclockwise from the x-axis
* shadow - draws a shadow beneath the pie (give a 3D feel)

```python
df_cons['Total'].plot(kind='pie', 
                      figsize = (15,6),
                      autopct='%1.1f%%',
                      startangle=90,
                      shadow=True)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # sets the pie chart to look like a circle
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each cont. w/ which to offset each wedge

df_cons['Total'].plot(kind='pie', 
                      figsize = (15,6),
                      autopct='%1.1f%%',
                      startangle=90,
                      shadow=True,
                      labels=None, #turn off labels
                      pctdistance=1.12, #ratio bet. center of each slice & start of text generated by autopct
                      colors=colors_list,
                      explode=explode_list) #explode lowest 3 conts.

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) #scale title up by 12% to match pctdist
plt.axis('equal') # sets the pie chart to look like a circle

# add legend
plt.legend(labels=df_cons.index, loc='upper left')
```

### Box Plots

to represent data distribution through 5 main dimensions:  

* minimum -> smallest num in the dataset
* 1st quartile -> mid num bet. the min. & the median
* 2nd quartile (median) -> mid num of the sorted dataset
* 3rd quartile -> mid num bet. median and max
* maximum -> highest num in the dataset

#### Subplots

untuk memplot bbrp plot dlm figure yg sama, misalnya melakukan perbandingan box plot scr berdampingan dgn line plot imigrasi China & India

```python
# General syntax:
fig = plt.figure() #create fig
ax = fig.add_subplot(nrows, ncols, plot_number) #create subplots
```

plot_number starts at 1, increments across rows first & has a maximum of nrows * ncols 

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, 1st plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, 2nd plot)

# Subplot 1: Box plot
df_CI.plot(kind='box', color='blue', vert=False, figsize=(20,6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China & India in 1980 - 2013')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df_CI.plot(kind='line', figsize=(20,6), ax=ax1) # add to subplot 2
ax1.set_title('Line Plots of Immigrants from China & India in 1980 - 2013')
ax1.set_xlabel('Number of Immigrants')
ax1.set_ylabel('Countries')
```

In the case when nrows, ncols and plot_number are all less than 10, a convenience exists, such that the a 3 digit number can be given instead, where the hundreds represent nrows, the tens represent ncols and the units represent plot_number. For instance:  

subplot(211)

produces a subaxes in a figure which represents the top plot (i.e. the first) in a 2 row by 1 column notional grid (no grid actually exists, but conceptually this is how the returned subplot has been positioned).

```python
# step 1: dapatkan dataset
# krn kt menggunakan hub antara years & total pop., kt akan mengubah years menjadi int.

# use the sum() method to get the total pop. per year
df_tot = pd.DataFrame(df[years].sum(axis=0))

# change the years to type int (useful for regression l8r)
df_tot.index = map(int,df_tot.index)

# reset the index to put in back in as a column in the df_tot df
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'total']

df_tot.head()
```

### Bubble Plots

variasi dari scatter plot yg menampilkan 3D data (x,y,z)  
data point diganti dgn bubbles  \
ukuran bubbles ditentukan oleh 3rd variable 'z' = bobot (weight)  
kt bs mengirimkan array / skalar ke keyword s to plot() -> bobot stiap titik

Compare Argentina & Brazil's immigration in 1980 - 2013 with a bubble plot as the normalised pop. value for every year

```python
# step 1: get data 
# convert the Years to type it & bring it in the df
df_t = df[years].T # transposed df

# cast the Years (index) to type int
df_t.index = map(int, df_t.index)

# label the index -- which will automatically be the column name when we reset index
df_t.index.name = 'Year'

# reset index to bring the Year in as a column
df_t.reset_index(inplace=True)

df_t.head()
```

in this case, we're using feature scaling -- membawa semua nilai ke dlm kisaran [0.1]  

rumus x' = x - xmin / xmax - xmin  

x = og, nilai max. dlm kumpulan data ke 1, dan nilai min = 0. other data pints are scaled between 0-1

```python
# step 2: create normalized weights

# normalize Brazil data
norm_brazil = (df_t['Brazil'] - df_t['Brazil'].min()) / (df_t['Brazil'].max() - df_t['Brazil'].min())

# normalize Argentina data
norm_argentina = (df_t['Argentina'] - df_t['Argentina'].min()) / (df_t['Argentina'].max() - df_t['Argentina'].min())
```

to plot 2 diff scatter plots in one plot, we can include the axes one plot into the other by passing it via the ax parameter  

we'll also pass in the weights using the s parameter.  
given that the normalized weights are bet. 0-1, they won't be visible on the plot. therefore, we will:  

* multiply weights by 2000 to scale it up on the graph, and,
* add 10 to compensate for the min. value ( which has a 0 weight & therefore scale with x 2000)

```python
# step 3: plot
# Brazil
ax0 = df_t.plot(kind='scatter',
                x='Year',
                y='Brazil',
                figsize=(14,8),
                alpha=0.5, #3D, transparency
                color='green',
                s=norm_brazil * 2000 + 10, # pass in weights
                xlim=(1975, 2015))

# Argentina
ax1 = df_t.plot(kind='scatter',
                x='Year',
                y='Argentina',
                alpha=0.5, #3D, transparency
                color='blue',
                s=norm_argentina * 2000 + 10,
                ax = ax0)

ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.set_ylabel('Number of Immigrants')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

bubble size suits how great the immigrating population is for that year. -- the bigger the bubble, the more immigrant is present in that year.  

from the above plot, we notice an improvement regarding immigration from Argentina during the Great Depression in 1998 - 2002.  

we can also observe a similar steady increase around the year 1985 to 1993.  
the fact states that Argentina had experience the G.D. from the year 1974 - 1990, before the G.D began in the year 1998 - 2002.  

on the same note, Brazil suffered from the Samba Effect, where the Brazil currency drops almost to 35% in 1999.  
there was a fear of financial crisis in South America because a lot of Southern America countries heavily depended on Brazil's  export industry.

the Brazil gov. then adopted some savings program, and the economy slowly recovers for years, which peaked on the increase in 2010. All of these circumstances were reflected in the immigration data
