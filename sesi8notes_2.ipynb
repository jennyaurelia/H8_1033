#### Maps with Markers

```python
df_ori = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_inci = df_ori
df_inci
```

```python
# get the first 100 crimes in the df
limit = 100
df_inci = df_inci.iloc[0:limit, :]
```

```python
df_inci.shape
```

```python
# SF latitude & longitude values
latitude = 37.77
longitude = -122.42
```

```python
# create map & display it
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sf_map
```

```python
# instantiate a feature group for the incidents in the df
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes & add each to the incident's feature group
for lat, lng, in zip(df_inci.Y, df_inci.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big u want the circle markers

            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
sf_map.add_child(incidents)
```

```python
# add pop up text
incidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_inci.Y, df_inci.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big u want the circle markers

            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop up text to each marker
latitudes = list(df_inci.Y)
longitudes = list(df_inci.X)
labels = list(df_inci.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sf_map)
    
# add incidents to map
sf_map.add_child(incidents)
```

solution 1: menghapus marker lokasi & menambahkan teks ke marker lingkaran

```python
# create map & display it
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label in zip(df_inci.Y, df_inci.X, df_inci.Category):
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big u want the circle markers

            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6).add_to(sf_map)
        
sf_map
```

solution 2: mengelompokkan marker ke dlm cluster yg berbeda

ps: suka ini bgt!!

```python
from folium import plugins
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# instantiate a mark cluster obj. for the incidents in the df
incidents = plugins.MarkerCluster().add_to(sf_map)

# loop thru the df & add each data point to rhe mark cluster
for lat, lng, label, in zip(df_inci.Y, df_inci.X, df_inci.Category):
    folium.Marker(
        location=[lat,lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

sf_map
```
