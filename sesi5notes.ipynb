pencarian untuk game Baltimore dimana kedua tim mencetak lebih dari 100 point dan harus exclude duplicates

```python
df[
    (df['_iscopy'] == 0) &
    (df['pts'] > 100) &
    (df['opp_pts'] > 100) &
    (df['team_id'] == 'BLB')
]
```

```python
df[
    (df['fran_id'] == 'Warriors') &
    (df['year_id'] == 2015)
].groupby(['is_playoffs','game_result'])['game_id'].count()
```
